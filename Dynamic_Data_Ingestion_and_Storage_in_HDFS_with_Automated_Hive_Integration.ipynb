```python
#!/usr/bin/python  =using vm ware which is working fine , but the oython version in vmware is 2.4 so it doesn't allow fetching files from https
#so i have created a file in vmware and the path is also mentioned below code.
# -*- coding: utf-8 -*-

import os
import sys

# ------------------------
# Configuration
# ------------------------
local_csv = "/home/training/product_data.csv"  # <-- local CSV path
hdfs_csv_dir = "/product_data"
sqoop_hdfs_dir = "/sqoop_movies"

# ------------------------
# Step 0: Verify Local CSV
# ------------------------
print "Using local CSV file:", local_csv
if not os.path.exists(local_csv) or os.path.getsize(local_csv) == 0:
    print "Error: CSV file does not exist or is empty."
    sys.exit(1)
print "CSV is ready."

# ------------------------
# Step 1: Upload CSV to HDFS
# ------------------------
print "\nUploading CSV to HDFS..."
os.system("hadoop fs -rmr %s" % hdfs_csv_dir)
os.system("hadoop fs -mkdir -p %s" % hdfs_csv_dir)
os.system("hadoop fs -put %s %s" % (local_csv, hdfs_csv_dir))
os.system("hadoop fs -ls %s" % hdfs_csv_dir)

# ------------------------
# Step 2: Hive Table for Product Data
# ------------------------
print "\nCreating Hive table for product data..."
hive_sql = """
DROP TABLE IF EXISTS product_table;
CREATE TABLE product_table(
    id INT,
    product_name STRING,
    price FLOAT,
    city STRING
)
ROW FORMAT DELIMITED
FIELDS TERMINATED BY ','
STORED AS TEXTFILE;

LOAD DATA INPATH '%s/product_data.csv' INTO TABLE product_table;

SELECT * FROM product_table LIMIT 10;
""" % hdfs_csv_dir

ret = os.system("hive -e \"%s\"" % hive_sql)
if ret != 0:
    print "Hive product_table creation failed."
else:
    print "Hive product_table created successfully."

# ------------------------
# Step 3: Sqoop Import of Movies table
# ------------------------
print "\nImporting Movies table from MySQL using Sqoop..."
os.system("hadoop fs -rmr %s" % sqoop_hdfs_dir)

sqoop_cmd = (
    "sqoop import "
    "--connect jdbc:mysql://localhost:3306/training "
    "--username training "
    "--password training "
    "--table Movies "
    "--target-dir %s "
    "-m 1" % sqoop_hdfs_dir
)
ret = os.system(sqoop_cmd)
if ret != 0:
    print "Sqoop import failed."
else:
    print "Sqoop import completed successfully."

# ------------------------
# Step 4: Hive Table for Movies
# ------------------------
print "\nCreating Hive table for Movies data..."
hive_movies_sql = """
DROP TABLE IF EXISTS movies_table;
CREATE TABLE movies_table(
    movieid INT,
    movie_name STRING,
    release_date STRING,
    imdb_url STRING,
    unknown_genre INT,
    action INT,
    adventure INT,
    animation INT,
    children INT,
    comedy INT,
    crime INT,
    documentary INT,
    drama INT,
    fantasy INT,
    film_noir INT,
    horror INT,
    musical INT,
    mystery INT,
    romance INT,
    sci_fi INT,
    thriller INT,
    war INT,
    western INT
)
ROW FORMAT DELIMITED
FIELDS TERMINATED BY ','
STORED AS TEXTFILE;

LOAD DATA INPATH '%s/part-m-00000' INTO TABLE movies_table;

SELECT * FROM movies_table LIMIT 10;
""" % sqoop_hdfs_dir

ret = os.system("hive -e \"%s\"" % hive_movies_sql)
if ret != 0:
    print "Hive movies_table creation failed."
else:
    print "Hive movies_table created successfully."

print "\nAll tasks completed successfully!"
```
